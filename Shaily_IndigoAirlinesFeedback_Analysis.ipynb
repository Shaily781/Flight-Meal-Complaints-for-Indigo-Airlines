```python
import pandas as pd
df = pd.read_csv("indigo_meal_feedback_case_study.csv")
df.info()
```

```python
#Check size of data
df.shape
```

```python
#make a copy and drop customer_id duplicates
df_drop_dup = df.drop_duplicates(subset='customer_id', keep='first')
```

```python
#check new data size
df_drop_dup.shape
```

```python
df_copy = df
```

```python
#make meal type proper for all rows and save
df_copy.loc[:, 'meal_type'] = df_copy['meal_type'].str.title()
```

```python
df_copy
```

```python
#make routes upper case
df_copy.loc[:,'route'] = df_copy['route'].str.upper()
```

```python
#fill all null values of complaing flag
df_copy.loc[:,'complaint_flag'] = df_copy['complaint_flag'].fillna("Not Available")
```

```python
#correct complaint flag wrong data
df_copy.loc[:, 'complaint_flag'] = df_copy['complaint_flag'].replace('noo','No')
```

```python
#fill taste rating values if null
df_copy.loc[:, 'taste_rating'] = df_copy['taste_rating'].fillna(3)
```

**Part A: Understanding the Problem Based on Taste Rating**

1. How many passengers gave a taste rating less than 3?

```python
df_copy[df_copy['taste_rating'] < 3].shape[0]
```

2. Out of those, how many were served by each meal provider?

```python
df_copy[df_copy['taste_rating'] < 3]['meal_provider'].value_counts()
```

3. Which routes had feedback with taste < 3? Print the feedback IDs and routes.

```python
df_copy[df_copy['taste_rating'] < 3][['feedback_id','route']]
```

```python
df_copy[df_copy['taste_rating'] < 3]['route'].value_counts() # frequncy of the route issue
```

4. Print all feedback entries where both taste and freshness were rated ≤ 2.

```python
df_copy[(df_copy['taste_rating'] <=2) & (df_copy['freshness_rating'] <= 2)]
```
